# Filter and convert SNP matrixes

This script takes as input the sliced matrixes derived from the script "human-te-dynamics-svn/scripts/sync2matrix.py". The goals of the script are to:

* Remove all the positions for which we have less than 1000 entries in at least one sample (**coverage < 1000**)
* Remove all the positions which are not polymorphic (**major allele > 99%**)
* Convert the raw numbers into **frequencies** (rounded at 3 decimal places)

```python
import pandas as pd
```

This chunk takes the input matrix and filters for positions with coverage > 1000.

```python
matrix = pd.read_table('/Volumes/Temp1/rpianezza/TE/SNP/matrixes/MER2.matrix')

#matrix = df.iloc[:, [0,1,2,3,4]]

cov = []
index = []

for i in range(matrix.shape[0]):                       #iterate over rows
    cov = []
    for j in range(matrix.shape[1]):                   #iterate over columns
        value = matrix.loc[i][j]
        if (j!=0 and j!=1):                            # calculate coverage
            coverage = 0
            parsed = []
            k = 0
            for component in value.split(':'):
                if k != 4:                             # exclude 5th value (non defined bases)
                    parsed.append(int(component))
                k+=1
            coverage = sum(parsed)
            cov.append(coverage)
    if min(cov)>=1000:                                 # save index if min coverage is >1000
        index.append(i)

filtered_raw = matrix.iloc[index, :]                   # filter the initial matrix
filtered = filtered_raw.reset_index(drop=True)
```

The chunk below calculate percentages of each base (included D) and remove all the monomorphic positions.

```python
cov = []
perc = []
index = []

for i in range(filtered.shape[0]):                    #iterate over rows
    cov = []
    perc = []
    for j in range(filtered.shape[1]):                #iterate over columns
        value = filtered.loc[i][j]
        if (j!=0 and j!=1):                           # calculate coverage
            coverage = 0
            parsed = []
            k = 0
            for component in value.split(':'):
                if k!=4:                              # exclude 5th value (non defined bases)
                    parsed.append(int(component))
                k+=1
            coverage = sum(parsed)
            cov.append(coverage)
            u = 0
            for component in value.split(':'):
                if u!=4:
                    percentage = int(component)/coverage  # calculate percentages
                    perc.append(percentage)
                u+=1
    if max(perc)<=0.99:                               # save index if the major allele is <99%
        index.append(i)
        
filtered2_raw = filtered.iloc[index, :]               # filter again the dataset
filtered2 = filtered2_raw.reset_index(drop=True)
cols = filtered.columns.tolist()
```

The last chunk creates a new matrix containing frequencies instead of raw counts. It maintains the same format, with every frequency separated by ":" in each cell.

```python
frequency = pd.DataFrame()                         # create empty dataframe to fill

for i in range(filtered2.shape[0]):                # iterate over rows
    for j in range(filtered2.shape[1]):            # iterate over columns
        value = filtered2.loc[i][j]
        new_value = ""
        if (j!=0 and j!=1):                        
            coverage = 0
            parsed = []
            per = []
            for component in value.split(':'):
                parsed.append(int(component))
                coverage = sum(parsed)
            y = 0
            for component in value.split(':'):
                percentage = int(component)/coverage
                if y == 0:
                    new_value += (str(round(percentage, 3)))
                else:
                    new_value += (':'+str(round(percentage, 3)))
                y+=1
            frequency.loc[i,j]=new_value
        elif j==0:
            frequency.loc[i,j]=str(value)
        else:
            frequency.loc[i,j]=value
```

```python
frequency.columns=cols
frequency.to_csv("/Users/rpianezza/Desktop/MER2_frequency_matrix", sep = " ")
```
