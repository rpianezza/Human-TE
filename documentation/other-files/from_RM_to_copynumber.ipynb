```python
import pandas as pd
import csv
```

```python
RM = pd.read_csv('/Users/rpianezza/TE/T2T/RM_cutoff.csv', low_memory=False)
```

```python
subset = RM #RM.loc[:1000]
copynumbers = {}
tot = 0
te = ""
te_length = 0

for i in range(len(subset)):
    length = 0
    if subset.iloc[i]['matching_repeat'] != te:
        if i != 0:
            copynumbers.update({te:(tot/te_length)})
            te = subset.iloc[i]['matching_repeat']
        if subset.iloc[i]['strand'] == '+':
            length = abs(subset.iloc[i]['position_in_repeat_begin']-subset.iloc[i]['position_in_repeat_end'])
            te_length = length + subset.iloc[i]['position_in_repeat_left'] + subset.iloc[i]['position_in_repeat_begin']
            tot = length
        else:
            length = abs(subset.iloc[i]['position_in_repeat_left']-subset.iloc[i]['position_in_repeat_end'])
            te_length = length + subset.iloc[i]['position_in_repeat_left'] + subset.iloc[i]['position_in_repeat_begin']
            tot = length
    else:
        if subset.iloc[i]['strand'] == '+':
            length = abs(subset.iloc[i]['position_in_repeat_begin']-subset.iloc[i]['position_in_repeat_end'])
            tot = tot+length
        else:
            length = abs(subset.iloc[i]['position_in_repeat_left']-subset.iloc[i]['position_in_repeat_end'])
            tot = tot+length
    if i == len(subset):
        copynumbers.update({te:(tot/te_length)})
        
        
print(copynumbers)
```

```python
with open('/Users/rpianezza/TE/T2T/RM_complete/copynumber_RM.csv', 'w') as f:
    w = csv.writer(f)
    for row in copynumbers.items():
        w.writerow(row)
```
